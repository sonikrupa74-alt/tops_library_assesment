```python
# Q.1 Convert a 1D array to a 2D array with 2 rows

import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])

# Convert to 2 rows
# we can also use -1 when one dimention is unknown or whenever we wanted it to be calculated by itself. for ex: res = a.reshape(2,-1),means 2*? = 6 , 6/2=3 
res = a.reshape(2,3)

print(res)
```

```python
# Q.2 Get the common items between a and b 
import numpy as np

a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])

res = np.intersect1d(a, b)
# np.intersect1d() returns the unique common element between two arrays
print(res)
```

```python
#Q.3 Get all items between 5 and 10 from a
import numpy as np

a = np.array([2, 6, 1, 9, 10, 3, 27])

res = np.where((a >= 5) & (a <= 10))

print(a[res]) 
```

```python
# pandas 
# 1. Compute the minimum,25th percentile,median,75th and maximum
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50,60])

print("Min :",s.min())
print("25% :",s.quantile(0.25))
print("Median :",s.median())
print("75% :",s.quantile(0.75))
print("Max :",s.max()) 

```

```python
# 10 : Multiple functions plotting = concept not covered
import numpy as np
import matplotlib.pyplot as plt

# Create array from 0 to 6 with increment of 0.3
x = np.arange(0, 6.1, 0.3)

# Plot x, x², x³, x⁴ with given styles
plt.plot(x, x, "ro", label="x")
plt.plot(x, x**2, "bs", label="x²")
plt.plot(x, x**3, "g", label="x³")
plt.plot(x, x**4, ":", label="x⁴")

# Set axis limits
plt.xlim(0, 6)
plt.ylim(0, 125)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plots of x, x², x³, and x⁴")

# Show legend
plt.legend()

# Display plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100,000 random points with a normal distribution
x = np.random.randn(100000)

# Histogram with 10 bins
plt.hist(x, bins=10)
plt.title("Histogram with 10 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Histogram with 20 bins
plt.hist(x, bins=20)
plt.title("Histogram with 20 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Histogram with 50 bins
plt.hist(x, bins=50)
plt.title("Histogram with 50 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Observation
# 10 bins: Broad view; less detail.
# 20 bins: More detailed distribution.
# 50 bins: Very detailed; shows the bell-shaped normal distribution more clearly.
```

```python
# 13 

### Features of TensorFlow

# 1. Open Source = Free to use and developed by Google.
# 2. Easy Model Building = Supports creating and training deep learning and machine learning models.
# 3. High Performance = Runs efficiently on CPUs, GPUs, and TPUs.
# 4. Scalable = Can handle small projects as well as large-scale applications.
# 5. Automatic Differentiation = Automatically computes gradients for model training.
# 6. Keras Integration = Includes Keras API for building neural networks easily.
# 7. Cross-Platform Support = Works on Windows, Linux, macOS, Android, and iOS.
# 8. TensorBoard = Provides visualization tools for monitoring training and debugging models.
# 9. Model Deployment = Supports deployment using TensorFlow Serving, TensorFlow Lite, and TensorFlow.js.
# 10. Flexible Architecture = Supports various neural networks like ANN, CNN, RNN, and Transformer models.
```

```python
# 14 

# TensorFlow Limitations

# 1. Steep learning curve, especially for beginners.
# 2. Requires high computational power for training large models.
# 3. Consumes a large amount of memory.
# 4. Debugging can be difficult due to complex computation graphs.
# 5. Installation and setup may be complicated on some systems.
# 6. Training deep learning models can take a long time.
# 7. Code can become lengthy and complex for advanced projects.
# 8. Performance depends on the availability of powerful hardware such as GPUs or TPUs.
# 9. Frequent updates may introduce compatibility issues with older code.
# 10. Small machine learning projects may not need TensorFlow, making it unnecessarily heavy.
```

```python
# 15 

# Supervised Machine Learning

# Supervised machine learning is a type of learning in which the model is trained using labeled data. This means the input data is provided along with the correct output. The model learns the relationship between inputs and outputs and uses it to predict results for new data.

# Examples:

# * Email spam detection
# * House price prediction
# * Student result prediction

# Common Algorithms:

# * Linear Regression
# * Logistic Regression
# * Decision Tree
# * Random Forest
# * Support Vector Machine (SVM)
# * K-Nearest Neighbors (KNN)

# Unsupervised Machine Learning

# Unsupervised machine learning is a type of learning in which the model is trained using unlabeled data. The model tries to find hidden patterns, relationships, or groups in the data without knowing the correct output.

# Examples:

# * Customer segmentation
# * Market basket analysis
# * Anomaly detection

# Common Algorithms:

# * K-Means Clustering
# * Hierarchical Clustering
# * DBSCAN
# * Principal Component Analysis (PCA)

# Difference Between Supervised and Unsupervised Learning

# | Supervised Learning                    | Unsupervised Learning                            |
# | -------------------------------------- | ------------------------------------------------ |
# | Uses labeled data                      | Uses unlabeled data                              |
# | Predicts known outputs                 | Finds hidden patterns or groups                  |
# | Used for classification and regression | Used for clustering and dimensionality reduction |
# | Example: Spam detection                | Example: Customer segmentation                   |
```
